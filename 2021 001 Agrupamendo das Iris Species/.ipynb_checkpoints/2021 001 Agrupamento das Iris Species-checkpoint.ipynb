### CÓPIA DO DATASET ORIGINAL

```python
# Cópiar o dataset original para remover o campo Species

dd = dados.copy()
```

```python
# Retirar o campo Species
# Também poderiar usar o comando: dd = dd[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]

dd = dd.drop(['Id', 'Species'], axis = 1)
```

```python
# Verificar os dados do novo dataset

dd.head()
```

# MODELO

```python
# Importar o módulo de clusterização

from sklearn.cluster import KMeans
```

```python
# Criar o modelo de IA
kmeans = KMeans(n_clusters = 2)

# Treinar o modelo de IA com os dados
kmeans.fit(dd)
```

```python
# Resultado do modelo de IA treinado

kmeans.labels_
```

### Teste do cutuvelo

```python
# Fazer o teste do cutuvelo para verificar qual é o melhor valor para o parametro n_cluster

# Variável para guardar a métrica
wcss = []
# Criar 20 modelos de AI e guardar o valor da métrica 'inertia_'
for i in range(1, 21):
    # Criar o modelo de IA
    kmeans = KMeans(n_clusters = i)
    # Treina o modelo de IA com os dados
    kmeans.fit(dd)
    # Guarda o valor da métrica
    wcss.append(kmeans.inertia_)
```

```python
# Plot o teste do cotovelo
# Procurar o melhor valor de x para y tema maior variação

fig, ax = plt.subplots()
plt.title("Teste de cotovelo", fontsize=15)
plt.plot(wcss)

ax.set_ylabel("wcss")
ax.set_xlabel("cluster")
```

```python
# Confirmando o melhor valor do teste do cutuvelo

# Gerar os valores 
anterior = 1
for n, atual in enumerate(wcss):
    # Print nos valores
    print(n, '-', atual, '-', round( anterior / atual - 1) * 100)
    # Guardar o último valor
    anterior = atual
```

```python
# Identificado o valor do cluster 
# Criar o modelo IA e Fazer o treinamento final

kmeans = KMeans(n_clusters = 3)
kmeans.fit(dd)
```

```python
# Resultado do modelo de IA treinado

kmeans.labels_
```

# Decompor dos dados

```python
# Fazer a decomposição dos dados

# Importar o módulo para decomposição PCA
from sklearn.decomposition import PCA

# Criar o modelo de IA para decompor os dados em dois campos
pca = PCA(n_components=2)

# Treinar o modelo de IA com os dados
result = pca.fit_transform(dd)

# Criar dataframe do resultado
ddpca = pd.DataFrame(data = result, columns = ['pca1', 'pca2'])
```

```python
# Adicionar o resultado do agrupamento ou clusterização no dataset

ddpca['cluster'] = kmeans.labels_
```

```python
# Plot o scatter para verificar a clusterização

fig, ax = plt.subplots()
sn.scatterplot(x='pca1', y='pca2', data = ddpca, hue = 'cluster')
plt.title("Clusterização", fontsize=15)
```

# Comparação com os dados originais

```python
# Copiar o campo species para o dataset

ddpca['Species'] = dados['Species']
```

```python
# Plot agrupando pelo campo Species

fig, ax = plt.subplots()
plt.title("Classificação original", fontsize=15)
sn.scatterplot(x='pca1', y='pca2', data = ddpca, hue = 'Species')
```

# VALIDAÇÃO

### Esse tipo de modelo não supervisionado, não tem uma métrica para fazer a validação, pois ele não está fazendo a classificação e sim o agrupamento.
### Contudo, como os dados originais já estavam classificados, podemos usar esse parâmetro para analisar a precisão do modelo

```python
# Verificar os dados final

ddpca.head()
```

```python
# Agrupar para verificar a quantidade de cluster encontrado

ddpca.groupby(['cluster', 'Species'])['pca1'].count()
```

```python
# Criar um dicionário para trocar o valor numérico pela descrição 

dic = {}
# Loop para pegar os três valores 0, 1, 2
for cl in range(0, 3):
    # Guardar o primeiro indece 
    species = ddpca[ddpca['cluster'] == cl]['Species'].value_counts().index[0]
    # Adicionar o valor e a species no dicionário
    dic[cl] = species
    
print(dic)
```

```python
# Verificar ser se os valores estão sendo alterados

ddpca['cluster'].apply(lambda x: dic[x])
```

```python
# Criar um campo novo no dataset e guardar o novo valor

ddpca['Species ii'] = ddpca['cluster'].apply(lambda x: dic[x])
```

```python
# Verificar o dataset

ddpca.head()
```

```python
# Importar módulo IA para verificação da precisão do modelo de IA

from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
# Veririficar a precisão do modelo de IA

print(classification_report(ddpca['Species'], ddpca['Species ii']))
```

```python
# Verificar a precisão do modelo de IA

print(confusion_matrix(ddpca['Species'], ddpca['Species ii']))
```

# CONCLUSÃO

O modelo de agrupamento obteve 89% em média de acertos, o que é muito bom para este tipo de modelo de IA
