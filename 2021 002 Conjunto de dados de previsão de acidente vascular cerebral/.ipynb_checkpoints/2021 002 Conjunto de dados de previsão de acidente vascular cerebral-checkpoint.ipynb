# PRÉ-PROCESSAMENTO

```python
# Transformar os vamores texto em valores numéricos

#Pegar os valores das colunas numéricas
cl_num = set(dados.describe().columns)

# Pegar os valores de todas as colunas
cl = set(dados.columns)

# Tirar os colunas numéricas das colunas
cl_cat = cl - cl_num

# Verificar o resultado
print(cl_num)
print()
print(cl)
print()
print(cl_cat)
```

```python
# Reordenar o indice

dados.reset_index(inplace=True)
```

```python
# Usar a função do pandas para transformar uma coluna string em uma coluna numérica

dadosprocessados = dados
for cl in cl_cat:
    dadosprocessados[cl] = pd.Categorical(dadosprocessados[cl], categories = dadosprocessados[cl].unique(), ordered=True).codes
```

```python
# Verifiar o dataset

dadosprocessados
```

# MODELO

### REGRESSÃO LOGÍSTICA

```python
# Importar os módulos de IA

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# Dividir os dados em treino e teste

xtrain, xteste, ytrain, yteste = train_test_split(dadosprocessados.drop('stroke', axis = 1), dadosprocessados['stroke'], test_size = 0.3)
```

```python
# Criar o modelo de IA

modelo = LogisticRegression(max_iter=10000, tol=.001)
```

```python
# Treinar o modelo de IA

modelo.fit(xtrain, ytrain)
```

```python
# Criar as predições

predictions = modelo.predict(xteste)
```

```python
# Verificar as predições

predictions
```

```python
# Importar módulos de metricas

from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
# Criar as métricas

tbprecisao = []
tbprecisao.append(['REGRESSÃO LOGÍSTICA', classification_report(yteste, predictions), confusion_matrix(yteste, predictions)])
```

### SVM

```python
# Importar módulo de IA

from sklearn.svm import SVC
```

```python
# Criar o modelo de IA

modelosvm = SVC()
```

```python
# Treinar o modelo de IA

modelosvm.fit(xtrain, ytrain)
```

```python
# Criar as predições

predictions = modelosvm.predict(xteste)
```

```python
# Criar as métricas

tbprecisao.append(['SVM', classification_report(yteste, predictions), confusion_matrix(yteste, predictions)])
```

### GRIDSEARCH

```python
# Importar o módulo de IA

from sklearn.model_selection import GridSearchCV
```

```python
# Criar os parametros da IA

param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001]} 
```

```python
# Criar o modelo de IA

grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=2)
```

```python
# Treinar o modelo de IA

grid.fit(xtrain, ytrain)
```

```python
# Criar as predições

grid_predictions = grid.predict(xteste)
```

```python
# Criar as métricas

tbprecisao.append(['GRIDSEARCH',
                   classification_report(yteste, grid_predictions),
                   confusion_matrix(yteste, grid_predictions)])
```

### Decisio Tree Classifier

```python
# Importar o módulo de IA

from sklearn.tree import DecisionTreeClassifier
```

```python
# Criar o modelo de IA

arvore = DecisionTreeClassifier()
```

### RANDOM FOREST

```python
# Importar o módulo de IA

from sklearn.ensemble import RandomForestClassifier
```

```python
# Criar o modelo de IA

randomforest = RandomForestClassifier(n_estimators=200)
```

```python
# Treinar o modelo de IA

randomforest.fit(xtrain, ytrain)
```

```python
# Criar as predições

randomforest_predict = randomforest.predict(xteste)
```

```python
# Criar as métricas

tbprecisao.append(['RANDOMFOREST', classification_report(yteste, randomforest_predict),
                  confusion_matrix(yteste, randomforest_predict)])
```
